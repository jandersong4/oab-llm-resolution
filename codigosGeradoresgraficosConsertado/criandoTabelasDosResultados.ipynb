```python
import google.generativeai as genai
import pandas as pd
import numpy as np
import os
import time
import re
```

```python
def iterar_arquivos(caminho_checar):
    dataframes = []
    nomes = []
    caminhos = []
    for root, dir, files in os.walk(caminho_checar):
        if files:
            for file in files:
                caminho_arquivo = os.path.join(root, file)
                caminhos.append(caminho_arquivo)
                dataframe = pd.read_csv(caminho_arquivo)
                dataframes.append(dataframe)
                nomes.append(file)
    return dataframes, nomes, caminhos
```

```python
diretorio_atual = os.getcwd()
novo_diretorio = "Diretório com Tabelas de Acerto Todas as IAs"
novo_caminho = os.path.join(diretorio_atual, novo_diretorio)
onde_checar = "TodasAsRespostasSemX"
caminho_checar = os.path.join(diretorio_atual, onde_checar)
dataframes, nomes, caminhos = iterar_arquivos(caminho_checar)
```

```python
caminho = caminhos[0]
print(f"O caminho inicial é: {caminho}")
primeiro_diretorio = os.path.dirname(caminho)
diretorio = primeiro_diretorio
caminho_menos_arquivo = "placeholder"
lista_dataframes_concatenar = []
for df, nome, caminho in zip(dataframes, nomes, caminhos):
    caminho_ia = os.path.dirname(caminho) #diretorio atual
    caminho_prova = os.path.dirname(caminho_ia)
    ia = os.path.basename(caminho_ia)
    prova = os.path.basename(caminho_prova)
    if caminho_ia != diretorio: # se diretorio atual diferente do antigo
        print(f"O diretorio é: {diretorio}")
        tabela_tudo = pd.concat(lista_dataframes_concatenar).reset_index(drop=True)
        tabela_agrupada = tabela_tudo.groupby('model_name')['is_correct'].sum()
        ia2 = os.path.basename(diretorio)
        path_prova = os.path.dirname(diretorio)
        prova2 = os.path.basename(path_prova)
        nome_arq = "Tabela agrupada " + prova2 + " " + ia2 + ".csv"
        file_path = os.path.join(novo_caminho, nome_arq)
        print(f"O file path é: {file_path}\n\n")
        print("A tabela é:\n")
        print()
        print(tabela_agrupada)
        print()
        print("----------------------------------------------------------------")
        print("\n\n")
        os.makedirs(novo_caminho, exist_ok=True)
        tabela_agrupada.to_csv(file_path)
        lista_dataframes_concatenar = []
        lista_dataframes_concatenar.append(df)
        diretorio = caminho_ia
    else:
        lista_dataframes_concatenar.append(df)
tabela_tudo = pd.concat(lista_dataframes_concatenar).reset_index(drop=True)
tabela_agrupada = tabela_tudo.groupby('model_name')['is_correct'].sum()
ia2 = os.path.basename(diretorio)
path_prova = os.path.dirname(diretorio)
prova2 = os.path.basename(path_prova)
nome_arq = "Tabela agrupada " + prova2 + " " + ia2 + ".csv"
file_path = os.path.join(novo_caminho, nome_arq)
print(f"O file path é: {file_path}")
os.makedirs(novo_caminho, exist_ok=True)
tabela_agrupada.to_csv(file_path)
diretorio = caminho_ia
print(f"O diretorio é: {diretorio}\n\n\n")
print("A tabela é:\n")
print()
print(tabela_agrupada)
print()
print("----------------------------------------------------------------")
print("\n\n")
    
```
